```python
import pandas as pd

ds = pd.read_csv("car_data.csv")

ds.head()
```

```python
ds["Price"] = ds["Price"].str.replace(',','')

ds["Price"] = pd.to_numeric(ds["Price"])
ds.loc[:, "Price"] *= 0.012

ds.head()
```

```python
ds["Fuel_Tank_Capacity(L)"] = pd.to_numeric(ds["Fuel_Tank_Capacity(L)"])
ds.loc[:, "Fuel_Tank_Capacity(L)"] *= 0.264172

ds.head()
```

```python
ds = ds.rename(columns={"Fuel_Tank_Capacity(L)": "Fuel Tank Capacity(g)"})

ds.head()
```

```python
ds["Mileage(kmpl)"] = pd.to_numeric(ds["Mileage(kmpl)"])
ds.loc[:, "Mileage(kmpl)"] *= 2.35215
ds = ds.rename(columns={"Mileage(kmpl)": "Mileage(mpg)"})

ds.head()
```

```python
ds[ds["Mileage(kmpl)"] == "BS IV"].head()
```

```python
ds[ds["Mileage(kmpl)"] == "105 bhp @ 4400 RPM"].head()
```
